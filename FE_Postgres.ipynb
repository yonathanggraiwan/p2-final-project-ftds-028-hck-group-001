```python
import pandas as pd
import psycopg2 as db
import numpy as np
```

```python
conn_string="port=5432 dbname='postgres' host='localhost' user='postgres' password='postgres'"

conn=db.connect(conn_string)

df = pd.read_sql("select * from application_table", conn)
df2 = pd.read_sql("select * from credit_record", conn)
df2 = df2.drop(columns=['index_col'])

conn.close()
```

```python
df2['START_DATE'] = pd.to_datetime(df2['START_DATE'])
df2['SNAPSHOT_MONTH'] = pd.to_datetime(df2['SNAPSHOT_MONTH'])

df2['MONTHS_BALANCE'] = (df2['START_DATE'].dt.to_period('M') - df2['SNAPSHOT_MONTH'].dt.to_period('M')).apply(lambda x: x.n)
```

```python
df2.info()
```

```python
begin_month=pd.DataFrame(df2.groupby(["ID"])["MONTHS_BALANCE"].agg(min))
new_data=pd.merge(df2,begin_month,how="left",on="ID") 
new_data = new_data.rename(columns={
    'MONTHS_BALANCE_x': 'MONTHS_BALANCE',
    'MONTHS_BALANCE_y': 'MIN_BALANCE'
})
new_data = new_data.drop(columns=['MIN_BALANCE'])
new_data
```

```python
df2['credit_status'] = None
df2['credit_status'][df2['STATUS'] =='1']='Yes'
df2['credit_status'][df2['STATUS'] =='2']='Yes' 
df2['credit_status'][df2['STATUS'] =='3']='Yes' 
df2['credit_status'][df2['STATUS'] =='4']='Yes' 
df2['credit_status'][df2['STATUS'] =='5']='Yes' 
```

```python
df2
```

```python
count=df2.groupby('ID').count()
count['credit_status'][count['credit_status'] > 0]='NPL'  # no dia fraud
count['credit_status'][count['credit_status'] == 0]='Good Credit'  # yes dia lancar
count = count[['credit_status']]
new_data=pd.merge(new_data,count, how='inner', on='ID') 
new_data['target']=new_data['credit_status']
new_data.loc[new_data['target']=='NPL','target']=1
new_data.loc[new_data['target']=='Good Credit','target']=0
```

```python
df_new = pd.merge(df, new_data, on='ID')
df_new
```

```python
df_new.rename(columns={'CODE_GENDER':'gender','FLAG_OWN_CAR':'car','FLAG_OWN_REALTY':'realty',
                         'CNT_CHILDREN':'child_number','AMT_INCOME_TOTAL':'income',
                         'NAME_EDUCATION_TYPE':'education','NAME_FAMILY_STATUS':'marital_status',
                        'NAME_HOUSING_TYPE':'house_type','FLAG_EMAIL':'email',
                         'NAME_INCOME_TYPE':'occupation','FLAG_WORK_PHONE':'work_phone',
                         'FLAG_PHONE':'phone','CNT_FAM_MEMBERS':'family_size',
                         'MONTHS_BALANCE':'months_balance', 'STATUS':'status',
                         'FLAG_MOBIL':'mobile_phone'
                        },inplace=True)
df_new = df_new.drop(columns='OCCUPATION_TYPE')
df_new.columns = df_new.columns.str.lower()
df_new
```
