```python
begin_month=pd.DataFrame(df2.groupby(["ID"])["MONTHS_BALANCE"].agg(min))
begin_month=begin_month.rename(columns={'MONTHS_BALANCE':'begin_month'}) 
new_data=pd.merge(df2,begin_month,how="left",on="ID") 
new_data
```

```python
df2['credit_approval'] = None
df2['credit_approval'][df2['STATUS'] =='1']='Yes' 
df2['credit_approval'][df2['STATUS'] =='2']='Yes' 
df2['credit_approval'][df2['STATUS'] =='3']='Yes' 
df2['credit_approval'][df2['STATUS'] =='4']='Yes' 
df2['credit_approval'][df2['STATUS'] =='5']='Yes' 
```

```python
df2
```

```python
count=df2.groupby('ID').count()
count['credit_approval'][count['credit_approval'] > 0]='No'  # no dia fraud
count['credit_approval'][count['credit_approval'] == 0]='Yes'  # yes dia lancar
count = count[['credit_approval']]
new_data=pd.merge(new_data,count,how='inner',on='ID')
new_data['target']=new_data['credit_approval']
new_data.loc[new_data['target']=='No','target']=1
new_data.loc[new_data['target']=='Yes','target']=0
```

```python
new_data
```

```python
print(count['credit_approval'].value_counts())
count['credit_approval'].value_counts(normalize=True)
```

```python
df_new = pd.merge(df, new_data, on='ID')
df_new = df_new.drop(columns=['OCCUPATION_TYPE'])
df_new
```

```python
df_new.rename(columns={'CODE_GENDER':'gender','FLAG_OWN_CAR':'car','FLAG_OWN_REALTY':'realty',
                         'CNT_CHILDREN':'child_number','AMT_INCOME_TOTAL':'income',
                         'NAME_EDUCATION_TYPE':'education','NAME_FAMILY_STATUS':'marital_status',
                        'NAME_HOUSING_TYPE':'house_type','FLAG_EMAIL':'email',
                         'NAME_INCOME_TYPE':'occupation','FLAG_WORK_PHONE':'work_phone',
                         'FLAG_PHONE':'phone','CNT_FAM_MEMBERS':'family_size',
                         'MONTHS_BALANCE':'months_balance', 'STATUS':'status'
                        },inplace=True)

df_new.columns = df_new.columns.str.lower()
df_new
```
