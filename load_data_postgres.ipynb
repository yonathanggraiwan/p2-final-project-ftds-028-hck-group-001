```python
import pandas as pd
import psycopg2 as db
import numpy as np
```

```python
conn_string="port=5432 dbname='postgres' host='localhost' user='postgres' password='postgres'"

conn=db.connect(conn_string)

df = pd.read_sql("select * from application_record", conn)
df2 = pd.read_sql("select * from credit_history", conn)
```

```python
begin_month=pd.DataFrame(df2.groupby(["id"])["MONTHS_BALANCE"].agg(min))
begin_month=begin_month.rename(columns={'MONTHS_BALANCE':'begin_month'}) 
new_data=pd.merge(df2,begin_month,how="left",on="id") 
new_data
```

```python
df2['risk_category'] = None
df2['risk_category'][df2['STATUS'] =='1']='Yes' 
df2['risk_category'][df2['STATUS'] =='2']='Yes' 
df2['risk_category'][df2['STATUS'] =='3']='Yes' 
df2['risk_category'][df2['STATUS'] =='4']='Yes' 
df2['risk_category'][df2['STATUS'] =='5']='Yes' 
```

```python
new_data['STATUS'].value_counts()
```

```python
df2
```

```python
count=df2.groupby('id').count()
count['risk_category'][count['risk_category'] > 0]='Yes'  # yes dia fraud
count['risk_category'][count['risk_category'] == 0]='No'  # no dia lancar
count = count[['risk_category']]
new_data=pd.merge(new_data,count,how='inner',on='id')
new_data['target']=new_data['risk_category']
new_data.loc[new_data['target']=='Yes','target']=1
new_data.loc[new_data['target']=='No','target']=0
```

```python
count
```

```python
print(count['risk_category'].value_counts())
count['risk_category'].value_counts(normalize=True)
```

```python
df_new = pd.merge(df, new_data, on='id')
df_new
```

```python
df_new.rename(columns={'CODE_GENDER':'gender','FLAG_OWN_CAR':'car','FLAG_OWN_REALTY':'realty',
                         'CNT_CHILDREN':'child_number','AMT_INCOME_TOTAL':'income',
                         'NAME_EDUCATION_TYPE':'education','NAME_FAMILY_STATUS':'marital_status',
                        'NAME_HOUSING_TYPE':'house_type','FLAG_EMAIL':'email',
                         'NAME_INCOME_TYPE':'income_type','FLAG_WORK_PHONE':'work_phone',
                         'FLAG_PHONE':'phone','CNT_FAM_MEMBERS':'family_size',
                         'MONTHS_BALANCE':'months_balance', 'STATUS':'status',
                        'OCCUPATION_TYPE':'occupation'
                        },inplace=True)

df_new.columns = df_new.columns.str.lower()
df_new
```
